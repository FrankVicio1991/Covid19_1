```python
#Tomando datos mayores a 200.000 para graficar
infectados = Data_covid[Data_covid['NewConfirmed']>160000].sort_values('NewConfirmed', ascending=False)
```

```python
#utilizando los campos a graficar
plt.plot(infectados.Country, infectados.NewConfirmed)
#Titulo eje x
plt.xlabel("País")
#Titulo eje y
plt.ylabel("Nuevos Casos")
#Titulo
plt.title("Casos de Covid 19")

infectados
```

```python
#Tomando datos mayores a 25.000.000 para graficar
total_infectados = Data_covid[Data_covid['TotalConfirmed']>25000000].sort_values('TotalConfirmed', ascending=False)
#utilizando los campos a graficar
plt.plot(total_infectados.Country, total_infectados.TotalConfirmed / 10**6)
#Titulo eje x
plt.xlabel("País")
#Titulo eje y
plt.ylabel("Total Casos Confirmados")
#Titulo
plt.title("Casos de Covid 19")

total_infectados
```
